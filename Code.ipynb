```python
#Backward Bigram
import random
#How many f1 present in model
def counting_back(f1,model):
  var=0
  # print("f1:",f1)

  specimen=model.copy()
  # for x in specimen:
  #   print(x)
  while specimen:
    # print("\n")
    s1,s2=specimen.pop(0)
    # print("s1:",s1)
    # print("s2:",s2)
    if f1==s1  and s1!='e' or f1==s2:
      if s1!='s' and s2!='s':
        var=var+1
        # print("Match Found")
  return var


#Probability of words after value
def CFD_back(value,model):
  # print("\n")
  # c1,c2=value
  # for x in model:
  #   print(x)
  # print("C1:",c1)
  # print("c2:",c2)
  test1=model.copy()
  count=counting_back(value,model)
  # print("Count:",count)
  arr = [0 for j in range(len(model))]
  if count!=0:
    for j in range(len(model)):
      var=0
    
      test2=model.copy()
      p1,p2=test1.pop(0)
      # print("P1:",p1)
      # print("P2:",p2)
      if p2==value and p1!='s' and p1!='e': #p2 matched
        temp=p1 
        # print("Temp:",temp) #Obtain mext val
        for i in range(len(model)):
          q1,q2=test2.pop(0)
          if q2==value:
            d=0
            # print("First Passed")
            if q1==temp:
              # print("Sec Passed")
              var=var+1
      arr[j]=var/count
  # print("Arr:",arr)
    return arr
  else:
    return -1

#Creating tupples
def backward_bigram_func(poetry):
  #print(poetry)
  bigrams_back = list()
  w1='0'
  w2='0'
  temp='0'
  for i in range(len(poetry)-1):
    # print("\n")
    # print("Value of i:",i)
    w1 = poetry[i]
    w2 = poetry[i+1]
    # print("W1:",w1)
    # print("w2:",w2)
  
    if w1!='s':
      if w2=='e':
        temp=w1
      if w1=='e':
        #bigrams_back.pop(-1)
        bigrams_back.append(tuple((temp, w2)))
      if w2!='e':
        bigrams_back.append(tuple((w1, w2)))
        # print("Appended")
   
    if w2=='s':
      if i+2<len(poetry)-1:
        w3=poetry[i+2]
        bigrams_back.append(tuple((w1, w3)))
  return bigrams_back


back_bi=list()
back_bi=backward_bigram_func(token_poetry)
print("Token Poetry:",token_poetry)
# print("BackwardBigram:",back_bi)
for x in back_bi:
  print(x)
end_list=[0 for p in range(len(back_bi))]
end_list,size_end=ending_words(back_bi)
print("Size_end:",size_end)
start_list= [0 for j in range(len(back_bi))]
start_list,size=starting_words(back_bi)
back = open('/content/drive/My Drive/NLP/BackwardBigram.txt', "w")
for i in range(3): #Count of verses
  print("\n")
  back.write("\n")
  for i in range(4):
    verse=""
    verse_num=0
    verse_num=random.randint(7,10)
    #print("verse_num:",verse_num)
    for j in range(verse_num):#Number if words in verse
      if j==0:
        end_word=random.randint(0,size_end-1)
     
        print("End_list:",end_list[end_word])
        # p,printq=back_bi[end_list[end_word]]
        # temp1=back_bi[end_list[end_word]]
        p=end_list[end_word]
        # print("P:",p)
        # print("Q:",q)
        l=p
        verse=p+verse
      else:
        # print("l:",l)
        freq=CFD_back(l,back_bi)
        if freq!=-1:
          # print(freq)
          max_value=max(freq)
          max_index = freq.index(max_value)
          # print("Max_index:",max_index)
          # print("Next Word:",bigram_list[max_index])
          p,q=back_bi[max_index]
          temp=back_bi[max_index]
          # print("P:",p)
          # print("q:",q)
          verse=p+" "+verse
          l=p
        else:
          # print("Out of bound")
          ending=random.randint(0,size_end-1)
          temp1=end_list[ending]
          # print("Entered temp1:",temp1)
          if temp1!=0:
            verse=str(temp1)+" "+verse
    print("Verse:",verse)
    back.write(verse)
    back.write("\n")

```
